```python
import pandas as pd
import pickle

input_file = open('pivot_dict_allGenres.dat','rb')
pivot_dict = pickle.load(input_file)
input_file.close()
```

```python
def flatten_dict(df):
    col = df.columns

    num_row = len(df.Genre)
    num_col = len(col)
    flatten_array = []
    for r in range(0,num_row):
        origin = df.Genre[r]
        for c in range(1,num_col):
            dest = col[c].lower()
            count = df[col[c]][r]
            if count>0:
                flatten_array.append([origin, dest, count])

    flatten_df = pd.DataFrame(flatten_array,columns = ['Origin','Dest','Count'])
    return flatten_df
```

```python
flatten_dict(pivot_dict[1955])
```

```python
minYear = 1954
maxYear = 2010

writer = pd.ExcelWriter('flatten_shared_songs.xlsx', engine='xlsxwriter')
flatten_dictionary = {}
for year in range(minYear,maxYear+1):
    df = pivot_dict[year]
    flatten_df = flatten_dict(df)
    flatten_dictionary[year] = flatten_df
    flatten_df.to_excel(writer,str(year))
writer.save()

output_file = open('flatten_dict_allGenres.dat','wb')
pickle.dump(flatten_dictionary, output_file)
output_file.close()
```
