```python
import pandas as pd
import pickle

input_file = open('pivot_dict_allGenres.dat','rb')
pivot_dict = pickle.load(input_file)
input_file.close()
```

```python
df = pivot_dict[2007]
del df['HipHop']
rows = df.index
columns = df.columns[0:len(df.columns)]
df
```

```python
genres = df.Genre

df_byDecade = pd.DataFrame(0,index=rows, columns = columns)
df_byDecade['Genre'] = genres
df_byDecade
```

```python
input_file = open('acoustic_dict_allGenres.dat','rb')
acoustic_dict = pickle.load(input_file)
input_file.close()
```

```python
acoustic_dict[1954]
```

```python
columns = acoustic_dict[2007].columns
df_byDecade = pd.DataFrame(0,index=rows, columns = columns)
df_byDecade['Genre'] = genres
total_genre = pd.DataFrame(0, index = genres, columns = ["Total"])


minYear = 1954
maxYear = 2010

keys = ['1950s', '1960s', '1970s', '1980s', '1990s', '2000s']
current_key = keys[0]
acoustic_decade_dict = {}


for year in range(minYear,maxYear+1):
    print(str(year) + ' ' + current_key)
    if year == 1960:
        
        for g in df_byDecade.Genre:
            index = np.where(df_byDecade.Genre==g)[0]
            weight = total_genre[total_genre.index == g].values[0][0]
            if weight>0:
                df_byDecade.ix[df_byDecade.Genre==g]=df_byDecade[df_byDecade.Genre==g].drop('Genre', axis = 1)/weight
            df_byDecade.Genre = genres
        
        acoustic_decade_dict[current_key] = df_byDecade
        df_byDecade = pd.DataFrame(0,index=rows, columns = columns)
        df_byDecade['Genre'] = genres
        total_genre = pd.DataFrame(0, index = genres, columns = ["Total"])
        current_key = keys[1]
    if year == 1970:
        
        for g in df_byDecade.Genre:
            index = np.where(df_byDecade.Genre==g)[0]
            weight = total_genre[total_genre.index == g].values[0][0]
            if weight>0:
                df_byDecade.ix[df_byDecade.Genre==g]=df_byDecade[df_byDecade.Genre==g].drop('Genre', axis = 1)/weight
            df_byDecade.Genre = genres        
        
        acoustic_decade_dict[current_key] = df_byDecade
        df_byDecade = pd.DataFrame(0,index=rows, columns = columns)
        df_byDecade['Genre'] = genres
        total_genre = pd.DataFrame(0, index = genres, columns = ["Total"])
        current_key = keys[2]
    if year == 1980:

        for g in df_byDecade.Genre:
            index = np.where(df_byDecade.Genre==g)[0]
            weight = total_genre[total_genre.index == g].values[0][0]
            if weight>0:
                df_byDecade.ix[df_byDecade.Genre==g]=df_byDecade[df_byDecade.Genre==g].drop('Genre', axis = 1)/weight
            df_byDecade.Genre = genres
            
        acoustic_decade_dict[current_key] = df_byDecade
        df_byDecade = pd.DataFrame(0,index=rows, columns = columns)
        df_byDecade['Genre'] = genres
        total_genre = pd.DataFrame(0, index = genres, columns = ["Total"])
        current_key = keys[3]
    if year == 1990:

        for g in df_byDecade.Genre:
            index = np.where(df_byDecade.Genre==g)[0]
            weight = total_genre[total_genre.index == g].values[0][0]
            if weight>0:
                df_byDecade.ix[df_byDecade.Genre==g]=df_byDecade[df_byDecade.Genre==g].drop('Genre', axis = 1)/weight
            df_byDecade.Genre = genres
            
        acoustic_decade_dict[current_key] = df_byDecade
        df_byDecade = pd.DataFrame(0,index=rows, columns = columns)
        df_byDecade['Genre'] = genres
        current_key = keys[4]
    if year == 2000:
        
        for g in df_byDecade.Genre:
            index = np.where(df_byDecade.Genre==g)[0]
            weight = total_genre[total_genre.index == g].values[0][0]
            if weight>0:
                df_byDecade.ix[df_byDecade.Genre==g]=df_byDecade[df_byDecade.Genre==g].drop('Genre', axis = 1)/weight
            df_byDecade.Genre = genres
            
        acoustic_decade_dict[current_key] = df_byDecade
        df_byDecade = pd.DataFrame(0,index=rows, columns = columns)
        df_byDecade['Genre'] = genres
        total_genre = pd.DataFrame(0, index = genres, columns = ["Total"])
        current_key = keys[5]
    df = pivot_dict[year]
    
    #Add the total number of songs
    df_diag = pd.DataFrame([df[df.Genre[i].title()][i] for i in range(0,len(df.index))])
    df_diag = df_diag.set_index(df.Genre)
    df_diag.columns = ["Total"]
    total_genre=total_genre.add(df_diag, fill_value=0)
    
    
    acoustic_df = acoustic_dict[year]
    
    for g in acoustic_df.Genre:
        index = np.where(acoustic_df.Genre==g)[0]
        weight = df_diag[df_diag.index == g].values[0][0]
        df_byDecade.ix[df_byDecade.Genre==g]=df_byDecade[df_byDecade.Genre==g].add(weight*acoustic_df.ix[index].set_index(np.where(df_byDecade.Genre==g)[0]), fill_value=0)
        df_byDecade.Genre = genres
        
for g in df_byDecade.Genre:
    index = np.where(df_byDecade.Genre==g)[0]
    weight = total_genre[total_genre.index == g].values[0][0]
    if weight>0:
        df_byDecade.ix[df_byDecade.Genre==g]=df_byDecade[df_byDecade.Genre==g].drop('Genre', axis = 1)/weight
    df_byDecade.Genre = genres        

acoustic_decade_dict[current_key] = df_byDecade
```
