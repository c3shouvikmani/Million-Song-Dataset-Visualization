```python
import json
import pandas as pd
import pickle
input_file = open('pivot_dict_byDecade.dat','rb')
pivot_dict = pickle.load(input_file)
input_file.close()
```

```python
year = '1950s'
df = pivot_dict[year]
df = df.drop(['Genre'], 1)
df
```

```python
data = {
    'name' : list(df.columns),
    'color' : ["#9ACD32", "#377DB8", "#F5DEB3", "#EE82EE", "#40E0D0", "#FF6347",
                "#D8BFD8", "#D2B48C", "#4682B4", "#00FF7F", "#FFFAFA", "#708090",
                "#708090", "#6A5ACD", "#87CEEB"]
}
genre_color_df = pd.DataFrame(data)
genre_color_df
```

```python
matrix = []
matrix_np = df.as_matrix()
for vector in matrix_np:
    matrix.append(list(vector))
json.dumps(matrix)
```

```python
matrixFilePath = 'visualizations/data/matrix' + str(year) + '.json'
genresFilePath = 'visualizations/data/genres' + str(year) + '.csv'
with open(matrixFilePath, 'w') as f:
    json.dump(matrix, f)
genre_color_df.to_csv(genresFilePath, index=False)
```
