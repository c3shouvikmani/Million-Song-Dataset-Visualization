```python
import json
import pandas as pd
import pickle
input_file = open('pivot_dict_byDecade.dat','rb')
pivot_dict = pickle.load(input_file)
input_file.close()
```

```python
df = pivot_dict["2000s"]
df = df.drop(['Genre'], 1)
df
```

```python
genres = list(df.columns)
colors = ["#9ACD32", "#377DB8", "#F5DEB3", "#EE82EE", "#40E0D0", "#FF6347",
          "#D8BFD8", "#D2B48C", "#4682B4", "#00FF7F", "#FFFAFA", "#708090",
          "#708090", "#6A5ACD", "#87CEEB"]
genre_color_df = pd.DataFrame(genres, colors)
genre_color_df.columns = ["name"]
genre_color_df.set_index = "color"
genre_color_df
```

```python
matrix = []
matrix_np = df.as_matrix()
for vector in matrix_np:
    matrix.append(list(vector))
json.dumps(matrix)
```

```python
with open('visualizations/matrix.json', 'w') as f:
    json.dump(matrix, f)
genre_color_df.to_csv("visualizations/genres.csv")
```
